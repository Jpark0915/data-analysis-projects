```r
library("readxl")
library("ggplot2")
library("tidyr")
library("dplyr")
```

```r
df = read_excel("/kaggle/input/mpg-car-mileage/carmileage.xlsx")
```

```r
head(df)
df = df[,-1]
head(df)
```

```r
summary(df)
```

```r
attach(df)
boxplot(VOL, main = "Volume (Cubic Feet)")
boxplot(HP, main = "Engine Horsepower")
boxplot(MPG, main = "Miles Per Gallon")
boxplot(SP, main = "Top Speed")
boxplot(WT, main = "Weight (lbs)")
```

```r
cor.test(MPG,VOL, method = "pearson")
```

```r
plot(VOL,MPG, main = "Scatterplot of Volume and MPG", xlab = "volume", ylab = "MPG", pch = 19)
abline(lm(MPG~VOL), col = 4, lwd = 3)
```

```r
cor.test(MPG,WT, method = "pearson")
plot(WT,MPG, main = "Scatterplot of WT and MPG", xlab = "WT", ylab = "MPG", pch = 19)
abline(lm(MPG~WT), col = 4, lwd = 3)
```

```r
fit <- lm(MPG~VOL + HP + WT + SP, data=df)
summary(fit)
```

```r
res <- resid(fit)
plot(fitted(fit), res)
abline(0,0)
```

```r
anova(fit)
```

- Individually, all 4 feautres had an impact on predicting MPG 
- Collectively, WT, HP, SP were significant predictors to MPG 
  - Volume was not a significant predictor. This could be due to the multicolinearity effect
